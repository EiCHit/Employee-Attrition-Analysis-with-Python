```python
# How many employees left the company?
attrition_counts = df['Attrition'].value_counts()
print(attrition_counts)

sns.countplot(x='Attrition', data=df)
plt.title('Employee Attrition Count')
plt.show()
```

```python
# What’s the attrition rate by gender?
gender_attrition = pd.crosstab(df['Gender'], df['Attrition'], normalize='index') * 100
print(gender_attrition.round(2))

gender_attrition.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Attrition Rate by Gender (%)')
plt.ylabel('Percentage')
plt.show()
```

```python
# Does job satisfaction affect attrition?
job_satisfaction = df.groupby('Attrition')['JobSatisfaction'].mean()
print(job_satisfaction)

sns.boxplot(x='Attrition', y='JobSatisfaction', data=df)
plt.title('Job Satisfaction by Attrition')
plt.show()
```

```python
# Which departments have the most attrition?
dept_attrition = df[df['Attrition'] == 'Yes']['Department'].value_counts()
print(dept_attrition)

sns.countplot(x='Department', data=df[df['Attrition'] == 'Yes'], order=dept_attrition.index)
plt.title('Attrition by Department')
plt.show()
```

```python
# Does working overtime affect attrition?
overtime_attrition = pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100
print(overtime_attrition.round(2))

sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Attrition by Overtime')
plt.show()
```

```python
# Relationship between Monthly Income and Attrition
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income vs Attrition')
plt.show()
```
